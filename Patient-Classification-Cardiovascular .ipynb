```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(random_state=1)  
lr.fit(Xtrain, ytrain)
```

<h1><center>Part B - using sklearn pipeline </center></h1>

### 14) (a) Read in heart.csv in to a dataframe called df; (b) Split the data into Xtrain, Xtest, ytrain, ytest - with 20% in test, and random_state=1; (c) Create two lists called numeric_features (with age, trestbps, chol, thalach, oldspeak), and categorical_features (with sex, cp, fbs, restecg, exang, slope, ca, thal)

```python
#read data 
df = pd.read_csv('heart.csv')

#split target and features
X = df.drop(['target'], axis=1)
y = df['target']

#split Xtrain, Xtest, ytrain, ytest 
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.2, random_state=1)

### Following code to deal with SetttingWithCopyWarning, and ensure we are working with a copy of the data and not a view
Xtrain = Xtrain.copy()
Xtest = Xtest.copy()
ytrain = ytrain.copy()
ytest = ytest.copy()

#create numeric and categorial features
numeric_features = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
categorical_features = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal']
```

### 15) Create a pipeline called "numeric_transformer" with a StandardScaler step called "ss" (using the same parameters used in Part A above)

```python
from sklearn.pipeline import Pipeline
numeric_transformer = Pipeline(steps=[('ss', StandardScaler(with_mean=True, with_std=True))])
```

### 16) Create a pipeline called "categorical_transformer" with a OneHotEncoder step called "ohe" (using the same parameters used in Part A above)

```python
categorical_transformer = Pipeline(steps=[('ohe', 
                                   OneHotEncoder(sparse=False, dtype=int, handle_unknown='error', categories='auto'))])
```

### 17) Create a column transformer called "preprocessor" with two transformers: (a) the first transformer called "num" which uses the numeric_transformer you defined above on the numeric_features; and (b) the second transformer called "cat" which uses the categorical_transformer defined above on the categorical_features

```python
from sklearn.compose import ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)])
```

### 18) Create a pipeline called "clf" with two steps: (a) the first step called "pp" which invokes the preprocessor you defined above; and (b) the second step called "lr" which involkes a logisitc regression model  (using the same parameters used in Part A above)

```python
clf = Pipeline(steps=[('pp', preprocessor),
                      ('lr', LogisticRegression(random_state=1))])
```

### 19) Fit the clf pipeline to the training data

```python
clf.fit(Xtrain, ytrain)
```

### 20) Predict and Evaluate clf pipeline on Xtest (same results as in Part A above)

```python
ypred = clf.predict(Xtest)
print (metrics.accuracy_score(ytest, ypred))
```
